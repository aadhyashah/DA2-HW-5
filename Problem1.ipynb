```python
import torch

x1 = torch.tensor(2.0, requires_grad=True)
x2 = torch.tensor(3.0, requires_grad=True)
x3 = torch.tensor(1.5, requires_grad=True)
x4 = torch.tensor(0.5, requires_grad=True)

u1 = x1 * x2              
u2 = u1 + x3              
u3 = u2 ** 2              
u4 = torch.exp(x4)        
u5 = u3 / u4              
u6 = x1 + x4              
u7 = torch.sin(u6)        

f = u5 + u7             

print("u1 =", u1.item())
print("u2 =", u2.item())
print("u3 =", u3.item())
print("u4 =", u4.item())
print("u5 =", u5.item())
print("u6 =", u6.item())
print("u7 =", u7.item())
print("f  =", f.item())
```

```python
f.backward()

print("\nGradients:")
print("df/dx1 =", x1.grad.item())
print("df/dx2 =", x2.grad.item())
print("df/dx3 =", x3.grad.item())
print("df/dx4 =", x4.grad.item())
```
